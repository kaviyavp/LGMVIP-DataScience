```python
import cv2
```

```python
img=cv2.imread("C:/Users/DELL/Downloads/test.jpg")
```

```python
cv2.imshow("Test", img)
cv2.waitKey(0)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.title('Nature')
plt.axis('off')
plt.imshow(img_rgb)
```

```python
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gray_img
```

```python
plt.title('Nature')
plt.axis('off')
plt.imshow(gray_img)
```

```python
inverted_img=255-gray_img
plt.title('Nature')
plt.axis('off')
plt.imshow(inverted_img)
```

```python
blurred=cv2.GaussianBlur(inverted_img, (21, 21), 0)
plt.title('Nature')
plt.axis('off')
plt.imshow(blurred)
```

```python
inverted_blurred=255-blurred
pencil_sketch=cv2.divide(gray_img, inverted_blurred, scale=256.0)
cv2.imshow("Sketch",pencil_sketch)
cv2.waitKey(0)
```
